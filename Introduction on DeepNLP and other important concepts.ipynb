# Introduction and why are we using Seq2Seq instead of BOW



1. It is important to design an end to end deep learning model since it often performs better than traditional deep learning models. Seq2Seq or sequence to sequence is an end to end deep learning model. Seq2seq is one of the most cutting edge and most powerful models which we have right now for nlp. 
2. An end to end deep learning model is preferred since there is only one model which is trained on all the modules completely. So there is not one deep learning model trained on something and the other deep learning model trained on some other module. This results in training the deep learning models trained separately and so the model overall will not be able to perform optimally. Example - If we have to pass an information via various ppl, some ppl will modify and the final meaning by the end is lost. So the models might be very good at what they do but when they perform together as a system, they fail. An end to end deep learning model will take care of everything from start to end.
